```python
from bs4 import BeautifulSoup as bs
import pandas as pd
import requests
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
```

## NASA Mars News

```python
# Splinter set up
executable_path = {'executable_path': ChromeDriverManager().install()}

browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Visit url: https://redplanetscience.com/
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
html = browser.html
soup = bs(html, 'html.parser')
soup
```

```python
# Scrape the Mars News Site and for the latest News Title 
newsTitle = soup.find('div', {'class': 'content_title'}).get_text()

#newsTitle = soup.find('div', class_ = 'content_title').get_text()
print(newsTitle)
```

```python
# Scrape the Mars News Site and for the Paragraph Text
newsPara = soup.find('div', {'class': 'article_teaser_body'}).get_text()
print(newsPara)
```

## Mars Hemispheres

```python
executable_path = {'executable_path': ChromeDriverManager().install()}

browser = Browser('chrome', **executable_path, headless=False)
url = 'https://marshemispheres.com/'
browser.visit(url)
```

```python
html = browser.html
hemiSoup = bs(html, 'html.parser')
hemiSoup
```

```python
# Storing for main url appending
mainHemiURL = 'https://marshemispheres.com/'
mainHemiURL
```

```python
# Locate links of hemispheres 'div', {'class': 'item'}
hemiURL = hemiSoup.find_all('div', {'class': 'item'})
hemiURL
```

```python
# Create dictionary to store urls
hemiURLDict = []

for hemi in hemiURL:
    hemisURL = hemi.find('a')['href']
    hemiURLDict.append(hemisURL)

print(hemiURLDict)
```

```python
hemiIMGurl = []
for hemimg in hemiURLDict:
    hemiURL = mainHemiURL + hemimg
    print(hemiURL)
```
